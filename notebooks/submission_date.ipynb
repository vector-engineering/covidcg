```python
# OK, we have to exclude invalid dates. let's do it by regex

valid_dates = df['date'].str.match('2020-([01]){1}([0-9]){1}-([0-9]){1,2}')
valid_dates
```

```python
df_filt = df.loc[valid_dates, :].reset_index(drop=True)
df_filt
```

```python
collection_dates = df_filt['date'].apply(pd.to_datetime)
collection_dates
```

```python
# Date to integer
collection_dates.astype(int)
```

```python
# Convert submission dates
submission_dates = df_filt['date_submitted'].apply(pd.to_datetime)
submission_dates
```

```python
# Get time difference
date_diff = submission_dates.astype(int) - collection_dates.astype(int)
date_diff
```

```python
date_diff_delta = date_diff.apply(pd.to_timedelta)
date_diff_delta
```

```python
date_diff_delta.mean()
```
